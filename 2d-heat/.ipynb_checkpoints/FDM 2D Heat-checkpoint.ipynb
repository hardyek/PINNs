<h4>2D Heat Equation</h4>

$$\frac{\partial u}{\partial t} = \alpha âˆ‡^{2}u$$

Code from: https://levelup.gitconnected.com/solving-2d-heat-equation-numerically-using-python-3334004aa01a

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

print("2D heat equation solver")

plate_length = 10
max_iter_time = 10

alpha = 0.5
delta_x = 1

delta_t = (delta_x ** 2)/(4 * alpha)
gamma = (alpha * delta_t) / (delta_x ** 2)

# Initialize solution: the grid of u(k, i, j)
u = np.empty((max_iter_time, plate_length, plate_length))

# Initial condition everywhere inside the grid
u_initial = 10.0

# Boundary conditions
u_top = 0.0
u_left = 0.0
u_bottom = 0.0
u_right = 0.0

# Set the initial condition
u.fill(u_initial)

# Set the boundary conditions
u[:, (plate_length-1):, :] = u_top
u[:, :, :1] = u_left
u[:, :1, 1:] = u_bottom
u[:, :, (plate_length-1):] = u_right

def calculate(u):
    for k in range(0, max_iter_time-1, 1):
        for i in range(1, plate_length-1, delta_x):
            for j in range(1, plate_length-1, delta_x):
                u[k + 1, i, j] = gamma * (u[k][i+1][j] + u[k][i-1][j] + u[k][i][j+1] + u[k][i][j-1] - 4*u[k][i][j]) + u[k][i][j]

    return u

def plotheatmap(u_k, k):
    # Clear the current plot figure
    plt.clf()

    plt.title(f"Temperature at t = {k*delta_t:.3f} unit time")
    plt.xlabel("x")
    plt.ylabel("y")

    # This is to plot u_k (u at time-step k)
    plt.pcolormesh(u_k, cmap=plt.cm.jet, vmin=0, vmax=100)
    plt.colorbar()

    return plt

# Do the calculation here
u = calculate(u)

def animate(k):
    plotheatmap(u[k], k)

anim = animation.FuncAnimation(plt.figure(), animate, interval=1, frames=max_iter_time, repeat=False)
anim.save('2Dheat.gif')
print("Done!")
```

I will take this data using a 100 x 100 plate and over a maximum of 100 time steps. I will also use 10 different initial condition functions outlined below along with the boundary conditions (Dirichlet homogeneous boundary conditions).

$$ \text{Boundary Conditions} $$
$$ u(0,y,t) = u(a,y,t) = 25, 0 \le x \le 100$$
$$ u(x,0,t) = u(x,b,t) = 25, 0 \le y \le 100$$

$$ \text{Initial Conditions} $$
$$\begin{align}
u(x,y,0) &= f(x,y) = \sin(\pi x)\sin(\pi y) \\
\end{align}$$

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
import csv

def calculate(u):
    for k in range(0, max_iter_time-1, 1):
        for i in range(1, plate_length-1, delta_x):
            for j in range(1, plate_length-1, delta_x):
                u[k + 1, i, j] = gamma * (u[k][i+1][j] + u[k][i-1][j] + u[k][i][j+1] + u[k][i][j-1] - 4*u[k][i][j]) + u[k][i][j]

    return u

def normalize_2d(matrix):
    diff = matrix.max() - matrix.min()
    matrix = ((matrix - matrix.min())/diff)
    return matrix

plate_length = 100
max_iter_time = 100

x = np.linspace(0, 1, plate_length)
y = np.linspace(0, 1, plate_length)
t = np.linspace(0, 1, max_iter_time)
X, Y = np.meshgrid(x, y)

alpha = 1
delta_x = 1


delta_t = (delta_x ** 2)/(4 * alpha)
gamma = (alpha * delta_t) / (delta_x ** 2)

u = np.empty((max_iter_time, plate_length, plate_length))

rows = [
    ['x','y','t','u_t']
]

def normalize_2d(matrix):
    diff = matrix.max() - matrix.min()
    matrix = (((matrix - matrix.min())*100)/diff)
    return matrix

# Define the initial condition functions with random parameters
x0 = 0.5   # x-coordinate of the center of the Gaussian
y0 = 0.5   # y-coordinate of the center of the Gaussian
sigma = 0.2 # standard deviation of the Gaussian
u_initial = np.exp(-((X-x0)**2 + (Y-y0)**2)/(2*sigma**2))

u[0,:,:] = u_initial

u_top = 0.0
u_left = 0.0
u_bottom = 0.0
u_right = 0.0

u[:, (plate_length-1):, :] = u_top
u[:, :, :1] = u_left
u[:, :1, 1:] = u_bottom
u[:, :, (plate_length-1):] = u_right

u = calculate(u)

for t_step in range(0, max_iter_time, 1):
    for x_step in range(0, plate_length, delta_x):
        for y_step in range(0, plate_length, delta_x):
                arr = [x[x_step],y[y_step],t[t_step],u[t_step,y_step,x_step]]
                rows.append(arr)

with open('2DHeatData.csv', mode='w', newline='') as file:

    writer = csv.writer(file)
    writer.writerows(rows)

print('CSV file written successfully.')

def plotheatmap(u_k, k):
    # Clear the current plot figure
    plt.clf()

    plt.title(f"Temperature at t = {k*delta_t:.3f} unit time")
    plt.xlabel("x")
    plt.ylabel("y")

    # This is to plot u_k (u at time-step k)
    plt.pcolormesh(u_k, cmap=plt.cm.jet, vmin=0, vmax=1)
    plt.colorbar()

    return plt

def animate(k):
    plotheatmap(u[k], k)

anim = animation.FuncAnimation(plt.figure(), animate, interval=10, frames=max_iter_time, repeat=False)
anim.save('2Dheat.gif')
print("Done!")
```
